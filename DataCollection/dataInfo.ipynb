```python
import pandas as pd

# Load the sampled dataset
df_sampled = pd.read_csv('sampled_processed_data.csv')

# Print basic information about the dataset
print("Dataset Info:")
print(df_sampled.info())

# Print the first few rows of the dataset
print("\nFirst 5 rows of the dataset:")
print(df_sampled.head())

# Print summary statistics of the dataset
print("\nSummary statistics:")
print(df_sampled.describe())

# Print the distribution of reviews per product (asin)
print("\nDistribution of reviews per product (asin):")
print(df_sampled['asin'].value_counts())

# Print the distribution of ratings
print("\nDistribution of ratings:")
print(df_sampled['rating'].value_counts(normalize=True))

# Print the distribution of reviews per user (user_id)
print("\nDistribution of reviews per user (user_id):")
print(df_sampled['user_id'].value_counts())
```
